```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
```

```python
import warnings
warnings.filterwarnings('ignore') 
```

```python
col_names = ['span','matches', 'innings', 'notout','runs', 'highscore', 'average', 'century', 'half_century','zeroes','profile']
data=pd.read_csv('/home/kanisk/Desktop/test_cricket',sep='\t',encoding='utf-8', index_col=0)
data.columns = col_names
```

```python
data.head()
```

```python
data.drop(["profile"], axis = 1, inplace = True) 
```

```python
data.drop(["span"], axis = 1, inplace = True) 
```

```python
data
```

```python
data.dtypes
```

```python
%%time
fig=sns.catplot(x='matches', y='runs', data=data,alpha=1.0)
plt.xticks([0,20,40,60,80,100,120,140,160,180,200])
plt.xlabel('Matches Played', fontsize=18)
plt.ylabel('Runs scored', fontsize=16)
plt.title('Striplot'+'(default)')
```

```python
%%time
fig=sns.catplot(x='matches', y='runs', data=data,kind='swarm',alpha=1.0)
plt.xticks([0,20,40,60,80,100,120,140,160,180,200])
plt.xlabel('Matches Played', fontsize=18)
plt.ylabel('Runs scored', fontsize=16)
plt.title('Swarmplot')
```

```python
%%time
sns.catplot(x='century', y='runs', data=data,kind="violin")
plt.xticks([0,4,8,12,16,20,24,28,32,36])
plt.xlabel('Centuries', fontsize=18)
plt.ylabel('Runs scored', fontsize=16)
plt.title('Violinplot')
```

```python
%%time
sns.catplot(x='half_century', y='runs', data=data,kind="boxen")
plt.xticks([0,8,16,24,32,40,48,56,64,72,80])
plt.xlabel('Half Centuries', fontsize=18)
plt.ylabel('Runs scored', fontsize=16)
plt.title('Boxenplot')
```

```python
%%time
sns.catplot(x='half_century', y='runs', data=data,kind="point",alpha=0.8)
plt.xlabel('Innings', fontsize=18)
plt.ylabel('Zeroes', fontsize=16)
plt.title('Pointplot')
```

```python
%%time
sns.catplot(x='innings', y='zeroes', data=data,kind="bar",alpha=0.8)
plt.xticks([30,60,90,120,150,180,210])
plt.xlabel('Innings', fontsize=18)
plt.ylabel('Zeroes', fontsize=16)
plt.title('Barplot')
```

```python
%%time
sns.catplot(x='half_century', data=data,kind="count",alpha=0.90)
plt.xticks([0,4,8,12,16,20,24,28,32,36,40])
plt.title('Countplot')
```
